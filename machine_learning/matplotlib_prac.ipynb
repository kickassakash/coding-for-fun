```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
plt.grid()
plt.plot([1,2,3,4],[4,3,2,1])
plt.xlabel("idk man")
plt.ylabel("ok man")
plt.title("first plot")
plt.show()
```

```python
plt.figure(1)
plt.subplot(211)#211 stands for - our figure has 2 rows 1 column and part/index 1 
plt.plot([1,2,3,4])#if not mentioned [1,2,3,4] is considered as y-axis points
plt.subplot(212)#212 stands for - our figure has 2 rows 1 column and part/index 2
plt.plot([1,2,3,4],[4,3,2,1])#first [1,2,3,4] is x axis and [4,3,2,1] is y-axis
```

```python
def square(n):
    return n**2

def root(n):
    return math.sqrt(n)

def linear(n):
    return n

t1 = np.arange(0,1,.2)
plt.sqr(t1,square(t1),'b--')
plt.plot(t1,root(t1),'g^')
plt.plot(t1,linear(t1),'rs')#t1=xaxis,func(t1) = yaxis, "b--" -> blue dotted lines
plt.grid()
plt.grid()
```
